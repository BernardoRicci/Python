I'm going to generate N points on a 2D plane, then I'll take only ones inside the circle of radius 1. After that I'll calculate Ï€ multiplying for 4 the number of points in circle and dividing by the number of total points. After that I'll plot it.

```python
import random
N=10**6
n=0
xin=[]  #for point inside the circle
yin=[]
xout=[]  #for point outside the circle
yout=[]
for i in range(N):
  x=random.uniform(-1,1)
  y=random.uniform(-1,1)
  if (x**2+y**2 <=1):
    n+=1
    xin.append(x)
    yin.append(y)
  else:
    xout.append(x)
    yout.append(y)
pi=(4*n)/N
print("N=%d, n=%d ,pi=%.6f" %(N,n,pi))

# Import modules 
import matplotlib.pyplot as plt
import math

plt.title("pi with MC method")
plt.scatter(xin,yin, color="green", marker=".")
plt.scatter(xout,yout, color="black", marker=".")

plt.show()
  
  
```
