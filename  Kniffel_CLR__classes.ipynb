Now we create a class to simulate a dice

```python
class Die:
  def __init__(self, number_dice, sides=6):
    #initialize a die, default 6 sides
    self.number_dice = number_dice
    self.sides = sides

  def roll(self):
    self.roll=r.randint(1,self.sides)
    return self.roll

  def __str__(self):
    info='(number of dice selected: {0}, result of dice: {1})'.format(str(self.number_dice), str(self.roll))
    return info

k=Dice(3)
print(k)
```

Class to simulate Kniffel game

```python
class Kniffel:
  def __init__(self):
    pass

  all_events = {'ks': "Kleine strasse",
              'gs': "Grosse strasse",
              'kn': "Kniffel",
              'fh': "Full house",
              '4k': "Viererpatsch",
              '3k': "Dreierpatsch"}


  def kniffel_roll():
    number_dice = Dice(5)
    hand=Dice.roll()

    return hand


  def check_score(hand):
    counts = [hand.count(n) for n in set(hand)]
    counts.sort()
 
    event = None   

    if counts == [1, 1, 3]:
        event = '3k'
    elif counts == [2, 3]:
        event = 'fh'
    elif counts == [1, 4]:
        event = '4k'
    elif len(counts) == 1:
        event = 'kn'
    elif sorted(hand) == [1, 2, 3, 4, 5]:
        event = 'ks'
    elif sorted(hand) == [2, 3, 4, 5, 6]:
        event = 'gs'

    return event


  def update_remaining_events(event, remaining_events):
    remaining_events.pop(event, None)


  def event_info(event, hand, remaining_events, all_events):
    if event is not None:
      try:
          print("{0}!".format(remaining_events[event]), hand)
      except:
          print("Too bad, you already scored a {0} previously!".format(all_events[event].lower()), hand)
    else:
      print("Better luck next time...", hand)



  N_games = 0
  all_rolls = []
  Total_games = int(1)

  while N_games < Total_games:
    remaining_events = dict(all_events)
    rolls = 0

  while len(remaining_events):
      rolls += 1
      hand = kniffel_roll()
      event = check_score(hand)
      #event_info(event, hand, remaining_events, all_events)
      update_remaining_events(event, remaining_events)
    
  N_games += 1
  all_rolls.append(rolls)
```
